```python
import seaborn as sb
d=sb.load_dataset("iris")
d.head()
```

```python
sb.relplot(x="sepal_length",y="sepal_width",data=d,hue="petal_length",kind="line")
```

```python
sb.relplot(x="sepal_length",y="petal_width",data=d,hue="species",col="species",kind="line")
```

```python
sb.relplot("total_bill","tip",data=df,style="smoker",size="size")
```

```python
sb.relplot("total_bill","tip",data=df,style="smoker",size="size",col="time",legend="full")
```
