## Representing a sine wave
A=sine(wt)
a=Amplitude
W=frequency
t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,1)
Amplitude=np.sin(Time)
pl.grid()
pl.plot(Time,Amplitude)
pl.Title("sine wave A=sin(wt)")
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl
```

```python
import pandas as pd
df=pd.read_csv('income.csv')
df
```

```python
import matplotlib.pyplot as pl
years=df.columns[2::]
Alaska_income=df.iloc[1,2:]
pl.plot(years,Alaska_income)
pl.title("income of Alaska from 2005 to 2013")
pl.xlabel("years")
pl.ylabel("Income of Alaska")
pl.savefig("income of Alaska from 2005 to 2013.jpg")
pl.show()
```

```python
# Plot for all states income from 2005 to 2013
## x-axis=>years
## Y-axis=>all states income
import matplotlib.pyplot as pl
years=df.columns[-5::]
allstate_income=df.iloc[0:,2:]
pl.plot(years,allstate_income)
pl.title("all states income 2005 to 2013")
pl.xlabel("years")
pl.ylabel("income")
pl.savefig("all states income 2005 to 2013.jpg")
pl.show()
```
