```python
import seaborn as sb
sb.set(style="darkgrid")
tips=sb.load_dataset("tips")
tips.head()
```

```python
help(sb.factorplot)
```

```python
sb.factorplot(x="total_bill",y="day",data=tips)
```

```python
sb.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=tips)
```

```python
sb.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=tips,dodge=True)
```

```python
from numpy import median
sb.factorplot(x="total_bill",y="day",hue="sex",markers=['x','s'],data=tips,dodge=True,estimator=median)
```

```python
sb.factorplot("sex","tip",hue="day",data=tips)
```

```python
help(sb.pointplot)
```

```python
sb.pointplot(x="day",y="tip",data=tips)
```

```python
sb.pointplot(x="day",y="tip",data=tips,hue="sex")
```

```python
sb.pointplot(x="time",y="tip",hue="sex",order=["Dinner","Lunch"],data=tips)
```

```python
sb.pointplot(x="time",y="tip",hue="sex",order=["Dinner","Lunch"],data=tips,color='r')
```

```python
sb.pointplot(x="time",y="tip",hue="sex",order=["Dinner","Lunch"],data=tips,palette="Dark2_r")
```

```python
sb.pointplot(x="time",y="total_bill",data=tips,hue="sex",markers=['x','s'],linestyles=["-","-."])
```

```python
sb.pointplot(x="tip",y="day",data=tips,hue="sex",join=False,dodge=True)
```

```python
sb.pointplot(x="tip",y="day",data=tips,hue="sex",join=False,dodge=True,estimator=median)
```

### Colors in seaborn

```python
import seaborn as sb
sb.palplot(sb.color_palette())
```

```python
c=sb.color_palette("RdBu")
sb.palplot(c)
```

```python
c=sb.color_palette("RdBu",n_colors=10)
sb.palplot(c)
```

```python
c=sb.color_palette("PuBu_r",15)
sb.palplot(c)
```

```python
sb.palplot(sb.color_palette("ch:3"))
```

```python
sb.palplot(sb.color_palette("ch:3,r=-1,l=-1"))
```

```python
sb.set()
c=sb.cubehelix_palette()
c
```

```python
sb.palplot(c)
```

```python
sb.palplot(sb.cubehelix_palette(reverse=True))
```

```python
sb.palplot(sb.cubehelix_palette(start=2,rot=-1))
```

```python
sb.palplot(sb.cubehelix_palette(start=2,rot=-1,reverse=True))
```

```python
import matplotlib.pyplot as plt
sb.heatmap(sb.cubehelix_palette())
plt.show()
```

```python
import numpy as np
x=np.arange(16).reshape(4,4)
c=sb.cubehelix_palette()
sb.heatmap(x,cmap=c)
```

```python
x=np.arange(16).reshape(4,4)
c=sb.cubehelix_palette(dark=-1)   # to show color
sb.heatmap(x,cmap=c)
```

```python
x=np.arange(16).reshape(4,4)
c=sb.cubehelix_palette(dark=-1,light=1,as_cmap=True)   # to show color
sb.heatmap(x,cmap=c)
```
