```python
x=[1,2,3,4]
y=[5,4,3,8]
plt.plot(x,y)
```

```python
plt.plot(x,y,linewidth=1.5,markersize=10,color='green')
```

```python
plt.plot(x,y,linewidth=1.5,markersize=10,color='green')
plt.xlabel("x")
plt.ylabel("y")
plt.title("graphing")
plt.show()
```

```python
x1=[5,6,8,9]
y1=[3,2,6,88]
plt.bar(x1,y1)
```

```python
x1=[5,6,8,9]
y1=[1,2,3,4]
plt.bar(x1,y1)
plt.bar(x,y,color='red')
```

## Historical plot

```python
plt.hist(x)
```

```python
help(plt.hist)
```

```python
bins=[2,4,6,8,9,10]
plt.hist(x,bins)
```

```python
bins=[2,4,6,8,9,10]
plt.hist(x,bins,histtype='bar',orientation='horizontal')
```

```python
bins=[2,4,6,8,9,10]    # should be in increasing order
plt.hist(x,bins,histtype='bar',orientation='horizontal',rwidth=5,color='k')
```
