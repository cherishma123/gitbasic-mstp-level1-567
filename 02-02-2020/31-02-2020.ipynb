```python
arr=[]
rv=poisson(25)
for num in range(0,40):
    arr.append(rv.pmf(num))
prob=rv.pmf(28)
plt.plot(arr,linewidth=2.0)
plt.plot([28],[prob],'.',c='r')
plt.show()
    
```
